```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
candy_filepath = "../programacion/candy-data.csv"

candy_data = pd.read_csv(candy_filepath)
```

```python
candy_data
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=candy_data['sugarpercent'], y=candy_data['winpercent']) 
```

```python
sns.regplot(x=candy_data['sugarpercent'], y=candy_data['winpercent'])
```

```python
sns.scatterplot(x=candy_data['pricepercent'], y=candy_data['winpercent'], hue=candy_data['chocolate'])
```

```python
sns.lmplot(x='pricepercent', y='winpercent', hue='chocolate', data= candy_data) 
```

```python
sns.swarmplot(x=candy_data['chocolate'],
              y=candy_data['winpercent'])
```

```python
sns.distplot(a=candy_data['pricepercent'], kde=False)
```

```python
sns.kdeplot(data=candy_data['winpercent'], shade=True)
```

```python
sns.jointplot(x=candy_data['winpercent'], y=candy_data['sugarpercent'], kind="kde")
```

```python
chocolate = candy_data[candy_data["chocolate"] == 1]
nochocolate = candy_data[candy_data["chocolate"] == 0]
```

```python
plt.figure(figsize=(20,15))
sns.set_style("whitegrid")
sns.distplot(a=chocolate['winpercent'], label="con chocolate", kde=False)
sns.distplot(a=nochocolate['winpercent'], label="sin chocolate", kde=False)
plt.title("Histograma exito del dulce, si lleva chocolate o no")
plt.legend()
```
